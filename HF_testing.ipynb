## Compute the functionality
### Least Cost Distance

Functionality is compuated as:

$L2all\_Q = landmarks.similarities\_L2all \cdot g.qualities$

$hf = landmarks.similarities\_all2L \cdot L2all\_Q$

**COMMENT: is this correct?**

```python
L2all_Q = landmarks.similarities_L2all.dot(g.qualities)
hf = landmarks.similarities_all2L.dot(L2all_Q)
hf[hf == 0] = np.nan

g.plot(hf)
```

### RSP Free Energy Distance 

```python
import scipy.sparse as ss
from scipy.sparse import linalg

def finite_max(a):
			''' Returns the maximum of an array, excluding infinte values
			'''
			return np.amax(a[np.isfinite(a)])
```

```python
landmarks.distances_all2L = landmarks.free_energy_distances_to(destinations=landmarks.landmarks, beta=0.01)
landmarks.similarities_all2L = (finite_max(landmarks.distances_all2L) - landmarks.distances_all2L)/finite_max(landmarks.distances_all2L)
landmarks.similarities_all2L[landmarks.distances_all2L == np.inf] = 0
'''assumes that the L2all equals all2L'''
landmarks.similarities_L2all = landmarks.similarities_all2L.transpose()
```

```python
L2all_Q = landmarks.similarities_L2all.dot(g.qualities)
hf = landmarks.similarities_all2L.dot(L2all_Q)
hf[hf == 0] = np.nan

g.plot(hf)
```
